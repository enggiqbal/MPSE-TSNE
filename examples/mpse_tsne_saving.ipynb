Here I describe how to use and save the results from running mview.mpse_tsne.

```python
import sys
sys.path.insert(0,"../")
import mview
```

When running mview.mpse_tsne, it is enough to set save_results to True before running the code, and the data will be saved in the folder mview/temp. You will then want to move the results elsewhere. For example:

```python
results = mview.mpse_tsne('mnist', n_samples=500, digits=[1,7],
                        perplexity=30,
                        show_plots=True, verbose=2,
                        save_results=True)
```

If you go to 'MPSE-TSNE/mview/temp/', you'll find the following csv files:

embedding.csv  images_0.csv  images_1.csv  projection_0.csv  projection_1.csv

In addition, mview.mpse_tsne returns the usual mview.MPSE object, which contains all the data regarding the instance.

```python
print('number of samples:', results.n_samples)
print('number of perspectives:', results.n_perspectives)
print('shape of embedding:', results.embedding.shape)
for i in range(results.n_perspectives):
    print('shape of projection',i,':', results.projections[i].shape)
    print('shape of image',i,':',results.images[i].shape)
```
