```python
import numpy as np 
import pandas as pd 
import pylab as plt
import mview
```

```python
data = pd.read_csv('palmerpenguins.csv')
data.head()
```

```python
x1 = data[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].to_numpy()
x1 /= np.max(x1,axis=0)
gender_map = {'male': 0, 'female': 1}
x2 = np.array([gender_map[item] for item in data['sex'].to_list()]).reshape((-1,1))
labels, index_map = pd.factorize(data['species'])
```

```python
marker_map = {0: "^", 1: "o"}
markers = [marker_map[gender[0]] for gender in x2.tolist()]
```

```python
proj1 = np.array([[1,0,0],
                  [0,1,0]])
proj2 = np.array([[1,0,0],
                  [0,1,0]])
mv = mview.mpse_tsne([x1,x2], smart_init=True, initial_projections=[proj1,proj2],
                perplexity=50,iters=500,sample_colors=[labels,labels],show_plots=False)
```

```python
tab10 = {0: "red", 
        1: "blue",
        2: "orange",
        3: "tab:red",
        4: "tab:purple"}
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
y1 = mv.images[0,:,:]
y2 = mv.images[1,:,:]
m = ['^', 'o']
for ax,emb in zip([ax1,ax2],[y1,y2]):
    x , y = emb[:,0], emb[:,1]
    for px,py,m,c in zip(x,y,markers, labels):
        ax.scatter(py,px,marker=m,c=tab10[c], alpha=0.5)
    ax.set_xticks([])
    ax.set_yticks([])

fig.set_size_inches(8,6)
# fig.tight_layout()
plt.savefig("figs/penguins-enstsne.png")
```

```python
fig= plt.figure(figsize=(5,4))
ax = fig.add_subplot(1,1,1,projection='3d')

X = mv.embedding

# perspectives = list()
# for k in range(mv.n_perspectives):
#     Q = mv.projections[k]
#     q = np.cross(Q[0],Q[1])
#     perspectives.append(q)

# q = perspectives
# for k in range(len(q)):
#     ind = np.argmax(np.sum(q[k]*X,axis=1))
#     m = np.linalg.norm(X[ind])/np.linalg.norm(q[k])
#     ax.plot([0,m*q[k][0]],[0,m*q[k][1]],[0,m*q[k][2]],'--',
#             linewidth=4.5,
#             color='gray')

N = len(X)

x,y,z = X[:,0], X[:,1], X[:,2]
for px,py,pz,m,c in zip(x,y,z,markers, labels):
    ax.scatter3D(px,py,marker=m,c=tab10[c], alpha=0.9)

ax.grid(color='r')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
plt.setp(ax.spines.values(), color='blue')
# ax.axes.get_xaxis().set_visible(False)
fig.set_size_inches(8,8)
fig.savefig('figs/penguins-enstsne-3d.png')
```

```python
X = np.concatenate((x1,x2),axis=1)
# X = x1
from sklearn.manifold import TSNE 
Y = TSNE(perplexity=8,learning_rate='auto',init='pca').fit_transform(X)

fig, ax = plt.subplots()
x , y = Y[:,0], Y[:,1]
for px,py,m,c in zip(x,y,markers, labels):
    ax.scatter(px,py,marker=m,c=tab10[c], alpha=0.9)
ax.set_xticks([])
ax.set_yticks([])
fig.set_size_inches(8,8)
plt.savefig("figs/penguins_full_tsne.png")
```
