This notebook contains examples on how to easily use mview to run mpse-tsne.This is accomplished using the function mview.mpse_tsne, which is optimized to use mview.MPSE for solving mpse-tsne problems.

```python
import sys
sys.path.insert(0,"../")
import mview
```

By default, mpse_tsne solves the mpse-tsne problem with variable projections, using a combination of gradient descent schemes (full gradient with fixed projections, batch gradient w/ multiple batch sizes, full gradient with adaptive and fixed rates).

The only required argument is the data describing the distances. In the next few examples I'll show how to run the examples that have already been incorporated into the module, in which case a single string is necessary. In the file 'mpse_tsne_running.ipynb' I'll describe how to run the code for data not already incorporated into the module.

In addition to the dataset name (the only argument of mview.mpse_tsne, other variables that help describe the data can be added too (for examples, the number of samples to use is n_samples). Some variables only affect a particular dataset, but there is no danger in adding a variable that is not applicable.

The saved datasets so far are 'equidistant', 'disk', 'clusters', 'clusters2', '123', 'florence', 'credit', 'phishing', and 'mnist'.

Below is the 3-perspective, 2-cluster example. The dataset name is 'clusters'. You can specify the number of clusters and number of perspectives as shown below.

```python
mview.mpse_tsne('clusters', n_samples=500, n_clusters=2, n_perspectives=3, perplexity=200,
               show_plots=True, verbose=2)
```

The same cluster example using a different configuration is shown below.

```python
mview.mpse_tsne('clusters', n_samples=500, n_clusters=[3,4,2], 
                perplexity=100,
                show_plots=True, verbose=2)
```

Here is mnist, which has name 'mnist'. We can only specify the number of samples here.

```python
mview.mpse_tsne('mnist', n_samples=500, 
                perplexity=30,
                show_plots=True, verbose=2)
```

Here we show the 123 example, with name '123'

```python
mview.mpse_tsne('123', 
                perplexity=900,
                show_plots=True, verbose=2)
```

Here is the florence example:

```python
mview.mpse_tsne('florence', perplexity=20,
                show_plots=True, verbose=2)
```

Here is the basic disk examples having multiple perspectives:

```python
mview.mpse_tsne('disk', n_samples=500, n_perspectives=5,
                perplexity=300, 
                show_plots=True, verbose=2)
```

Here is the phishing data example

```python
mview.mpse_tsne('credit',
                perplexity=30, 
                show_plots=True, verbose=2)
```
