```python
from sklearn.manifold import TSNE
from PIL import Image

import numpy as np
import matplotlib.pyplot as plt
import os, os.path, sys

sys.path.insert(1,'/Users/vahanhuroyan/Documents/MPSE/MPSE-TSNE')

```

```python
imgs = np.array([0] * 16384)
path = "./coil-20-proc"
for f in os.listdir(path):
    imgs = np.vstack([imgs, np.array(np.concatenate(np.array(Image.open(os.path.join(path,f)))))])    
imgs = imgs[1:, :]
```

```python
print(imgs.shape)
```

```python
tsne = TSNE(n_components=2, perplexity=10.0, random_state=0)
X_transformed_tsne = tsne.fit_transform(imgs)
plt.plot(X_transformed_tsne[:, 0], X_transformed_tsne[:, 1], '.', alpha=0.5)
```

```python
tsne = TSNE(n_components=2, perplexity=30.0, random_state=0)
X_transformed_tsne = tsne.fit_transform(imgs)
plt.plot(X_transformed_tsne[:, 0], X_transformed_tsne[:, 1], '.', alpha=0.5)
```
